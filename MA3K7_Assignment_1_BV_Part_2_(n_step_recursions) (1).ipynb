**3-Step Recursion**

The following code is used to find the number of distinct bracelets with a 3-step recursion rule using base 10.

Result is 20


```python
#we initialise the bracelet using the first 3 values which will give a unique bracelet
bracelet = [2,6,3]

#allows us to generate the bracelet by looping until we get back to the 2 intial values used to generate the bracelet
looping = True

#generation of the bracelet
while looping == True:
  bracelet.append((bracelet[-1]+bracelet[-2]+bracelet[-3]) % 10)
  if bracelet[0] == bracelet[-3] and bracelet[1] == bracelet[-2] and bracelet[2] == bracelet[-1]:
    looping = False

#output of the bracelet and the "length" of the bracelet
print(f"current bracelet: {bracelet}")
print(f"current bracelet length: {len(bracelet)-3}")
```

**5-Step Recursion**

The following code is used to find the number of distinct bracelets with a 5-step recursion rule using base 10.

Result is 40

```python
#intialising 2d array with all 100000 possible initial values
list_of_bracelets = []

for i in range(10):
  for j in range(10):
    for k in range(10):
      for l in range(10):
        for m in range(10):
          list_of_bracelets.append([i,j,k,l,m])

#generating all possible bracelets from the 100000 possible initial values
for i in range(100000):
  looping = True

  while looping == True:
    list_of_bracelets[i].append((list_of_bracelets[i][-1]+list_of_bracelets[i][-2]+list_of_bracelets[i][-3]+list_of_bracelets[i][-4]+list_of_bracelets[i][-5]) % 10)
    if list_of_bracelets[i][0] == list_of_bracelets[i][-5] and list_of_bracelets[i][1] == list_of_bracelets[i][-4] and list_of_bracelets[i][2] == list_of_bracelets[i][-3] and list_of_bracelets[i][3] == list_of_bracelets[i][-2] and list_of_bracelets[i][4] == list_of_bracelets[i][-1]:
      looping = False
```

```python
for i in range(100000):
  list_of_bracelets[i].append(False)

counter = 0

#checking to see which bracelets are non-distinct
for i in range(100000):
  if list_of_bracelets[i][-1] == False:
    counter += 1
    list_of_bracelets[i][-1] = counter
    for j in range(100000):
      for k in range((len(list_of_bracelets[j]))-5):
        if list_of_bracelets[j][k] == list_of_bracelets[i][0] and list_of_bracelets[j][k+1] == list_of_bracelets[i][1] and list_of_bracelets[j][k+2] == list_of_bracelets[i][2] and list_of_bracelets[j][k+3] == list_of_bracelets[i][3] and list_of_bracelets[j][k+4] == list_of_bracelets[i][4]:
          list_of_bracelets[j][-1] = counter

print(counter)
```

**Function that Returns Number of Distinct Bracelets for Base-n**

While this function should work in theory, in practice it does not function very well for $n \ge 5$ since the time-complexity of the algorithm is exponential

```python
def n_step_recursion(n):

  #initialising the 2d array for all possible bracelets
  list_of_bracelets = []

  zero_vector = []
  for i in range(n):
    zero_vector.append(0)

  #generating all possible initial values for an n-step recursion
  for i in range(10**n):
    new_vector = zero_vector.copy()
    current_i = i
    for j in range(n):
      new_vector[-(j+1)] = current_i % 10
      current_i = int((current_i - (current_i % 10)) / 10)
    list_of_bracelets.append(new_vector)

  #generating all possible bracelets from the initial values
  for i in range(10**n):
    looping = True

    while looping == True:

      #finding the next number in the bracelet
      to_append = 0
      for j in range(n):
        to_append += list_of_bracelets[i][-(j+1)]
      list_of_bracelets[i].append(to_append % 10)

      #comparing the last n values to the intial values to determine if the bracelet has terminated or not
      matching = True
      for j in range(n):
        if list_of_bracelets[i][j] != list_of_bracelets[i][-n+j]:
          matching = False

      if matching == True:
        looping = False

  #finding all distinct bracelets
  for i in range(10**n):
    list_of_bracelets[i].append(False)

  counter = 0

  #checking to see which bracelets are non-distinct
  for i in range(10**n):
    if list_of_bracelets[i][-1] == False:
      counter += 1
      list_of_bracelets[i][-1] = counter
      for j in range(10**n):
        if list_of_bracelets[j][-1] == False:
          for k in range((len(list_of_bracelets[j]))-n):
            matching = True
            for l in range(n):
              if list_of_bracelets[j][k+l] != list_of_bracelets[i][l] :
                matching = False
                break

            if matching == True:
              list_of_bracelets[j][-1] = counter

  return counter
```
